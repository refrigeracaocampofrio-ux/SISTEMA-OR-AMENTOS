```python
# Ler arquivo schema.sql
schema_path = 'database/schema.sql'

try:
    with open(schema_path, 'r', encoding='utf-8') as f:
        schema_content = f.read()
    print(f'‚úÖ Arquivo {schema_path} lido com sucesso!')
    print(f'   Tamanho: {len(schema_content)} caracteres')
except FileNotFoundError:
    print(f'‚ùå Arquivo {schema_path} n√£o encontrado!')
    exit(1)
```

```python
# Conectar e executar
print('\nüöÄ Iniciando importa√ß√£o do schema...')
print('=' * 60)

try:
    # Conectar ao banco
    connection = mysql.connector.connect(
        host=DB_HOST,
        user=DB_USER,
        password=DB_PASSWORD,
        database=DB_DATABASE,
        ssl_disabled=False
    )
    
    cursor = connection.cursor()
    print('‚úÖ Conectado ao PlanetScale com sucesso!')
    print()
    
    # Executar schema.sql
    statements = schema_content.split(';')
    executed = 0
    
    for statement in statements:
        statement = statement.strip()
        if statement and not statement.startswith('--'):
            try:
                cursor.execute(statement)
                executed += 1
                if executed % 5 == 0:
                    print(f'  üìã {executed} comandos executados...')
            except Exception as e:
                if 'already exists' not in str(e):
                    print(f'‚ö†Ô∏è  Erro ao executar: {str(e)[:100]}')
    
    # Commit
    connection.commit()
    print()
    print('=' * 60)
    print(f'‚úÖ‚úÖ‚úÖ SCHEMA IMPORTADO COM SUCESSO! ‚úÖ‚úÖ‚úÖ')
    print(f'   Total de {executed} comandos executados')
    print()
    
    cursor.close()
    connection.close()
    
except mysql.connector.Error as err:
    print(f'‚ùå Erro de conex√£o: {err}')
except Exception as e:
    print(f'‚ùå Erro: {e}')
```
